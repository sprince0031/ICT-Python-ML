### Question 1: How is the body mass of penguins distributed? Are there different groups?
***Hint:*** think which plot is used for visualising frequency counts of a feature.

```python
# Explore question 1: Using histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='body_mass_g', alpha=0.7, bins=50) # Can also use matplotlib directly
plt.title('Body mass distribution')
plt.xlabel('Body mass (g)')
plt.ylabel('Frequency')
plt.show()
```

### Question 2: Are flipper length and body mass related?
***Bonus:*** try the scatter plot (using seaborn) here with an extra parameter, `hue` which can visualise clusters.

```python
# Explore question 2: using scatterplot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', alpha=0.7)
plt.title('Flipper length vs Body mass')
plt.xlabel('Flipper length (mm)')
plt.ylabel('Body mass (g)')
plt.show()
```

```python
# Bonus: group by class to see how data points are clustered together for the flipper - body mass relationship
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', alpha=0.7, hue='species') # Just add hue param with target class
plt.title('Flipper length vs Body mass')
plt.xlabel('Flipper length (mm)')
plt.ylabel('Body mass (g)')
plt.show()
```

### Question 3: How does body mass differ between three species of penguin?
***Hint:*** species is a categorical feature. Think which plot is suitable for visualising spread within cateogries.

```python
# Explore question 3: using box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='species', y='body_mass_g')
plt.title('Body mass data spread for different species')
plt.xlabel('Penguin species')
plt.ylabel('Body mass (g)')
plt.show()
```

```python
# Bonus: Violin plot for individual class distribution like in histogram
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='species', y='body_mass_g')
plt.title('Body mass data spread for different species')
plt.xlabel('Penguin species')
plt.ylabel('Body mass (g)')
plt.show()
```

### Question 4: How do all numerical features correlate with each other?
***Hint:*** read the question again ;)

```python
# Explore question 4: visualise correlation matrix with heatmap
corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12, 6))
sns.heatmap(data=corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Heatmap of Palmer Penguins dataset')
plt.show()
```
